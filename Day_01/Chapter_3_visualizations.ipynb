### Bucketed Histograms of numeric values

```python
grades = [83,95,91,87,70,0,85,82,100,67,73,77,0]
decile = lambda grade: grade // 10 * 10
```

```python
histogram = Counter(decile(grade) for  grade in grades)

plt.bar([x-4 for x in histogram.keys()],   # shift each bar by 4 
       histogram.values() # get correct values 
       ,8) #assign width of 8
plt.axis([-5,105 , 0 ,5]) # x-axis from -5 to 105 and y-axis from 0 to 5
plt.xticks([10*i for i in range(11)])
plt.ylabel('decile')
plt.xlabel('grades')
plt.title('Distribution of Exam 1 grades')
plt.show()
```

### Misleading Axis

```python
mentions = [500, 505]
years = [2012.6, 2013.6]

plt.bar([2012.6,2013.6] , mentions , .8 , align="center" )
plt.xticks(years)
plt.ylabel("# of times I heard someone say 'data science'")
# if you don't do this, matplotlib will label the x-axis 0, 1
# and then add a +2.013e3 off in the corner (bad matplotlib!)

plt.ticklabel_format(useOffset=False)

#misleading axis only shows part above 500

plt.axis([2012.5,2014.5,499,506])
plt.show()
```

```python
plt.bar([2012.6,2013.6] , mentions , .8 , align="center" )
plt.xticks(years)
plt.ylabel("# of times I heard someone say 'data science'")
# if you don't do this, matplotlib will label the x-axis 0, 1
# and then add a +2.013e3 off in the corner (bad matplotlib!)

plt.ticklabel_format(useOffset=False)

#adjusting the axis a little bit 
plt.axis([2011.5,2014.5,0,506])
plt.show()
```

### Line Charts
good for showing trends

```python
variance = [1, 2, 4, 8, 16, 32, 64, 128, 256]
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1]
```

```python
total_error = [x+y for x,y in zip(variance,bias_squared)]
```

```python
xs = [i for i,_ in enumerate(variance)]
# multiple calls to plot for ploting the multiple vars on the same plot 

plt.plot(xs , variance , 'g-' , label="variance")
plt.plot(xs , bias_squared , 'r-.' , label="bias^2")
plt.plot(xs, total_error , 'b:' , label="total error")

plt.legend(loc=9)
plt.xlabel('model complexity')
plt.title('bias variance tradeoff')

plt.show()
```

### Scatter Plots
good for showing relationships

```python
friends = [ 70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']
```

```python
plt.scatter(friends,minutes)
for label , friend_count, minute_count in zip(labels , friends,minutes):
    plt.annotate(label,
                xy=(friend_count,minute_count),
                xytext=(5,-5),
                textcoords='offset points') #slight offset for better visibility
    

plt.title("Daily Minutes vs. Number of Friends")
plt.xlabel("# of friends")
plt.ylabel("daily minutes spent on the site")
plt.show()
```

### comparable values 
If youre scattering comparable variables, you might get a misleading picture if you let
matplotlib choose the scale

```python
test_1_grades = [ 99, 90, 85, 97, 80]
test_2_grades = [100, 85, 60, 90, 70]
```

```python
plt.scatter(test_1_grades,test_2_grades)
plt.title('axis arent comparable')
plt.xlabel('test_1_grades')
plt.ylabel('test_2_grades')
plt.show()
```

```python
# but if we add a plt.axis('equal') we get a more accurate plot

plt.scatter(test_1_grades,test_2_grades)
plt.title('axis are comparable')
plt.xlabel('test_1_grades')
plt.ylabel('test_2_grades')
plt.axis('equal')
plt.show()
```

### For Further Exploration

- seaborn is built on top of matplotlib and allows you to easily produce prettier (and
more complex) visualizations.

- D3.js is a JavaScript library for producing sophisticated interactive visualizations for

worth your while to be familiar with it.

- Bokeh is a newer library that brings D3-style visualizations into Python.

- ggplot is a Python port of the popular R library ggplot2, which is widely used for
creating publication quality charts and graphics. Its probably most interesting if
youre already an avid ggplot2 user, and possibly a little opaque if youre not.
